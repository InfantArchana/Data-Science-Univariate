```python
dataset=pd.read_csv("kidney_disease.csv")
```

# Finding missing values in dataset

```python
dataset.isnull().sum()
```

# Seperate Quan and Qual using Function

```python
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for ColumnName in dataset.columns:
        if (dataset[ColumnName].dtypes=="O"):
            Qual.append(ColumnName)
        else:
            Quan.append(ColumnName)
    return Quan,Qual 
```

```python
Quan,Qual = QuanQual(dataset)
```

```python
print("Quantitative columns:", Quan)
print("Qualitative columns:", Qual)
```

```python
Quan, Qual = QuanQual(dataset)
```

```python
dataset[Quan]
```

# Replacing Quan Missing/Null values with SimpleImputer

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
imp.fit(dataset[Quan])
Quan_dataset = imp.fit_transform(dataset[Quan])
Quan_dataset
```

```python
Quan_dataset = pd.DataFrame(Quan_dataset, columns=Quan)
Quan_dataset
```

```python
Quan_dataset.isnull().sum()
```

# Replacing Qual Missing/Null values with SimpleImputer

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='Unknown')
imp.fit(dataset[Qual])
Qual_dataset = imp.fit_transform(dataset[Qual])
Qual_dataset
```

```python
Qual_dataset = pd.DataFrame(Qual_dataset, columns=Qual)
Qual_dataset
```

```python
Qual_dataset.isnull().sum()
```

# Combine Quan_dataset, Qual_dataset DataFrames together using pd.concat

```python
Preprocessed_dataset = pd.concat([Quan_dataset, Qual_dataset], axis=1)
Preprocessed_dataset
```

```python
Preprocessed_dataset.isnull().sum()
```

# Save the Preprocessed dataset in csv file

```python
Preprocessed_dataset.to_csv("Preprocessed_CKD.csv", index=False)
```

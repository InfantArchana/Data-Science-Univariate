Logic to separate Quantitative and Qualitative Automatically

```python
#Creating a empty list and check the conditions using for loop
Quan=[]
Qual=[]
for ColumnName in dataset.columns:
    print(ColumnName)
    if (dataset[ColumnName].dtypes=="O"):
        print("Qual")
    else:
        print("Quan")
```

# Logic to separate Quan and Qual Automatically

```python
#Using append() method - Appends a element to the end of the list

Quan=[]
Qual=[]
for ColumnName in dataset.columns:
    print(ColumnName)
    if (dataset[ColumnName].dtypes=="O"):  # "O" is Object Data type
        print("Qual")
        Qual.append(ColumnName)
    else:
        print("Quan")
        Quan.append(ColumnName)
```

# Logic to separate Quan and Qual Automatically using Function

```python
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for ColumnName in dataset.columns:
        #print(ColumnName)
        if (dataset[ColumnName].dtypes=="O"):
            #print("Qual")
            Qual.append(ColumnName)
        else:
            #print("Quan")
            Quan.append(ColumnName)
    return Quan,Qual                 #Calling QuanQual() function.
```

```python
QuanQual(dataset)
```

```python
Quan, Qual = QuanQual(dataset)
```
